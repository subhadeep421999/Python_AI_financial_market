```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('ai_financial_market_daily_realistic_synthetic.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data['Date']=pd.to_datetime(data['Date'])
```

```python
data['Company'].unique()
```

```python
data['Year']=data['Date'].dt.year
```

```python
data.isnull().sum()
```

### How much amount the companies spent on R & D.

```python
print("Company's spending for R & D in $Bn:")
RD= data.groupby('Company')['R&D_Spending_USD_Mn'].sum()
RD
```

```python
plt.bar(RD.index, RD.values, color=['cyan','black','magenta'])
plt.title('R&D Spending by the companies')
plt.xlabel('Company')
plt.ylabel('Amount in USD_$Bn')
plt.show()
```

### Revenue Earned by the companies

```python
print("Company's AI_Revenue_USD_Mn:")
rev=data.groupby('Company')['AI_Revenue_USD_Mn'].sum()
rev
```

```python
plt.bar(rev.index, rev.values, color=['orange','green','blue'],width=0.5)
plt.title('Revenue Earned by the companies')
plt.xlabel('Company')
plt.ylabel('Amount in USD_$Bn')
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.bar(RD.index, RD.values, color=['cyan','black','magenta'])
plt.title('R&D Spending by the companies')
plt.xlabel('Company')
plt.ylabel('Amount in USD_$Bn')

plt.subplot(1,2,2)
plt.bar(rev.index, rev.values, color=['orange','green','blue'],width=0.5)
plt.title('Revenue Earned by the companies')
plt.xlabel('Company')
plt.ylabel('Amount in USD_$Bn')

plt.show()
```

### Datewise Impact on the Stock

### Expenditure vs Revenue year-by-year

```python
spend=data.groupby('Year')['R&D_Spending_USD_Mn'].sum()
spend
```

```python
plt.plot(spend.index, spend.values, color='r')
plt.title('Combined R&D Spending Year by Year')
plt.xlabel('Year')
plt.ylabel('Amount in USD_$Mn')
plt.show()
```

```python
REV=data.groupby('Year')['AI_Revenue_USD_Mn'].sum()
REV
```

```python
plt.plot(REV.index, REV.values, color='g')
plt.title('Combined Revenue earned Year by Year')
plt.xlabel('Year')
plt.ylabel('Amount in USD_$Mn')
plt.show()
```

```python
plt.plot(spend.index, spend.values, color='r')
plt.plot(REV.index, REV.values, color='g')
plt.title('Combined Expenditure vs Revenue Year by Year')
plt.xlabel('Year')
plt.ylabel('Amount in USD_$Mn')
plt.legend(['Expenditure','Revenue'])
plt.show()
```
